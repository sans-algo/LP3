```python
# Recursive Fibonacci
steps_rec = 0
def fib_recursive(n):
    global steps_rec
    steps_rec += 1
    if n <= 1:
        return n
    return fib_recursive(n - 1) + fib_recursive(n - 2)

# Non-Recursive Fibonacci
def fib_non_recursive(n):
    steps = 0
    a, b = 0, 1
    series = []
    for i in range(n):
        series.append(a)
        a, b = b, a + b
        steps += 1
    return series, steps

# Main program
n = int(input("Enter n: "))

# Recursive method
print("\nRecursive Fibonacci series:")
for i in range(n):
    print(fib_recursive(i), end=" ")
print("\nStep count (Recursive):", steps_rec)

# Non-Recursive method
series, steps_nonrec = fib_non_recursive(n)
print("\nNon-Recursive Fibonacci series:")
print(*series)
print("Step count (Non-Recursive):", steps_nonrec)


```
