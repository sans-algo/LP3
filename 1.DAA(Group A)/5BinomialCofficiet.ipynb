```python
# Generate Binomial Coefficients using Dynamic Programming

def binomial_coeff(n, k):
    C = [[0]*(k+1) for _ in range(n+1)]

    for i in range(n+1):
        for j in range(min(i, k)+1):
            if j == 0 or j == i:
                C[i][j] = 1
            else:
                C[i][j] = C[i-1][j-1] + C[i-1][j]
    return C[n][k]

# Example
n, k = 5, 2
print(f"C({n},{k}) =", binomial_coeff(n, k))
```
