```python
import heapq

# Huffman coding
def huffman(characters, freq):
    heap = [[f, [ch, ""]] for ch, f in zip(characters, freq)]
    heapq.heapify(heap)

    while len(heap) > 1:
        low = heapq.heappop(heap)
        high = heapq.heappop(heap)
        for pair in low[1:]:
            pair[1] = '0' + pair[1]
        for pair in high[1:]:
            pair[1] = '1' + pair[1]
        heapq.heappush(heap, [low[0] + high[0]] + low[1:] + high[1:])

    return sorted(heapq.heappop(heap)[1:], key=lambda p: (len(p[-1]), p))

# Example
chars = ['A', 'B', 'C']
freq = [5, 7, 10]

codes = huffman(chars, freq)

print("Huffman Codes:")
for ch, code in codes:
    print(ch, "->", code)
```
