```python
# Job Sequencing using Greedy Method

jobs = [('A', 2, 100), ('B', 1, 19), ('C', 2, 27), ('D', 1, 25), ('E', 3, 15)]

# Step 1: Sort by profit descending
jobs.sort(key=lambda x: x[2], reverse=True)

# Step 2: Find max deadline and make slots
n = max(j[1] for j in jobs)
slot = [0]*(n+1)
profit = 0
res = []

# Step 3: Fill slots greedily
for job in jobs:
    id, d, p = job
    for j in range(d, 0, -1):
        if slot[j] == 0:
            slot[j] = 1
            res.append(id)
            profit += p
            break

# Step 4: Print result
print("Jobs done:", res)
print("Total Profit:", profit)
```
