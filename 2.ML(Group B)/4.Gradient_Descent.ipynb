```python


import numpy as np
import matplotlib.pyplot as plt

# Define function and its derivative
def f(x):
    return (x + 3)**2

def grad(x):
    return 2 * (x + 3)

# Initialize parameters
x = 2.0          # starting point
lr = 0.1         # learning rate
iterations = 20  # number of steps

# For visualization
x_history = [x]
y_history = [f(x)]

# Gradient Descent loop
for i in range(iterations):
    g = grad(x)
    x = x - lr * g
    x_history.append(x)
    y_history.append(f(x))
    print(f"Iter {i+1:02d}: x = {x:.4f}, f(x) = {f(x):.6f}")

# Analytical minimum at x = -3
print("\nMinimum occurs at x ≈ -3, f(x) ≈ 0")

# Create smooth curve for the function
x_curve = np.linspace(-6, 4, 100)
y_curve = f(x_curve)

# Plot function and descent steps
plt.figure(figsize=(8, 5))
plt.plot(x_curve, y_curve, label='f(x) = (x+3)²', color='blue')
plt.scatter(x_history, y_history, color='red', label='Descent steps')
plt.plot(x_history, y_history, '--r', alpha=0.6)

# Mark the minimum point
plt.scatter(-3, f(-3), color='green', s=80, label='Minimum at x=-3')

# Labels and title
plt.title('Gradient Descent Visualization')
plt.xlabel('x values')
plt.ylabel('f(x) values')
plt.legend()
plt.grid(True)
plt.show()
```
