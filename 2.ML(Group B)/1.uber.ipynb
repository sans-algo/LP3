```python
# Uber Price Prediction using Linear Regression & Random Forest

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler

# 1️⃣ Load dataset
df = pd.read_csv('uber.csv')
print(df.head())

# 2️⃣ Preprocess data
df = df.dropna()

# Select only numeric columns
df = df.select_dtypes(include=['float64', 'int64'])

# Avoid invalid data (like infinite values)
df = df.replace([np.inf, -np.inf], np.nan).dropna()

# 3️⃣ Handle outliers (IQR method)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

# 4️⃣ Split data into features and target
X = df.drop('fare_amount', axis=1)
y = df['fare_amount']

# 5️⃣ Scale the data (important for Linear Regression)
sc = StandardScaler()
X_scaled = sc.fit_transform(X)

# 6️⃣ Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 7️⃣ Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

# 8️⃣ Random Forest Regression
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

# 9️⃣ Evaluate models
def evaluate(y_true, y_pred, model_name):
    r2 = r2_score(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"{model_name} → R2: {r2:.3f}, RMSE: {rmse:.3f}")

evaluate(y_test, y_pred_lr, "Linear Regression")
evaluate(y_test, y_pred_rf, "Random Forest Regression")
```
